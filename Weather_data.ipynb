```python
from meteostat import Stations, Daily
import pandas as pd
import os

# Time range
start_date = '2018-01-01'
end_date = '2022-12-31'

# Directory to save CSV files
save_directory = "C:\\Users\\shume\\Desktop\\6501.81_Capstone Project\\Refrences Data Comparision\\weather_data"

# Check if the directory exists, if not, create it
if not os.path.exists(save_directory):
    os.makedirs(save_directory)

# Define the station ID for Beitbridge
station_id = '67991'

# Function to fetch and save weather data
def fetch_and_save_weather_data(station_id, start_date, end_date, save_directory):
    # Fetch all stations and then filter by station ID
    all_stations = Stations().fetch()
    station = all_stations[all_stations.index == station_id]
    station_name = station.iloc[0]['name']
    
    # Get data
    data = Daily(station, start=start_date, end=end_date)
    data = data.fetch()
    
    # Save to CSV
    csv_filename = os.path.join(save_directory, f"{station_name}_weather_data.csv")
    data.to_csv(csv_filename)
    print(f"Saved data for {station_name} to {csv_filename}")

# Fetch and save weather data for the station
fetch_and_save_weather_data(station_id, start_date, end_date, save_directory)
```
